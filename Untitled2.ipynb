```r
data <- data.frame(
  Month = c("January","February","March","April","May","June",
            "July","August","September","October","November","December"),
  Category_A = c(600,600,700,500,800,800,500,700,500,800,500,600),
  Category_B = c(700,800,500,600,400,400,700,500,600,400,700,800),
  Category_C = c(600,700,600,800,300,200,600,800,800,300,600,700)
)
```

```r
install.packages("reshape2")
library(ggplot2)
library(reshape2)
data_long <- reshape2::melt(data, id.vars = "Month")
ggplot(data_long, aes(x = Month, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(
    title = "Vertical Bar Chart – Monthly Activity",
    x = "Month",
    y = "Value"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
ggplot(data_long, aes(x = Month, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  coord_flip() +
  labs(
    title = "Horizontal Bar Chart – Monthly Activity",
    x = "Month",
    y = "Value"
  ) +
  theme_minimal()
```

```r
ggplot(data_long, aes(x = Month, y = value, fill = variable)) +
  geom_bar(stat = "identity", position = position_dodge()) +
  labs(
    title = "Grouped Bar Chart – Category Comparison",
    x = "Month",
    y = "Value",
    fill = "Category"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
data <- data.frame(
  category = c("A", "B", "C"),
  group1 = c(10, 20, 30),
  group2 = c(15, 25, 35)
)

library(ggplot2)
ggplot(data, aes(x = category)) +
  geom_bar(aes(y = group1), stat = "identity", position = "dodge", fill = "blue") +
  geom_bar(aes(y = group2), stat = "identity", position = "dodge", fill = "red") +
  labs(title = "Side-by-Side Bar Chart")
```

```r
data <- data.frame(
  category = c("A", "B", "C"),
  group1 = c(10, 20, 30),
  group2 = c(15, 25, 35)
)

library(ggplot2)
ggplot(data, aes(x = category)) +
  geom_bar(aes(y = group1, fill = "Group 1"), stat = "identity") +
  geom_bar(aes(y = group2, fill = "Group 2"), stat = "identity") +
  labs(title = "Stacked Bar Chart") +
  scale_fill_manual(values = c("blue", "red"))
```
